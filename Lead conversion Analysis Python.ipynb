```python
#Create a new column as index in the dataset
index = range(0, len(data))
data = pd.DataFrame(data,index=index)
data['index_column'] = data.index
data
```

```python
#Using loop change the values of ‘Select’ to ‘NaN’ of city column
for i in range(len(data['City'])):
        if data['City'][i]=='Select':
            data['City'][i] =np.nan
           
data.head(10)
```

```python
    # Replacing all blank values of DataFrame data.
data = data.replace(r'^\s*$', np.nan, regex=True)
data
```

```python
#Computing the average time spent on the website
Avg_time = data['Total Time Spent on Website'].mean()
print('Average time spent on website is', round(Avg_time, 3),'seconds')
```

```python
#Adding a new column 'Total Time Spent on Website in Minutes' to data.
data['Total Time Spent on Website in Minutes'] = data['Total Time Spent on Website']/60
data['Total Time Spent on Website in Minutes'] = data['Total Time Spent on Website in Minutes'].round(decimals = 3)
data
```

```python
#Count of people w.r.t total time spent on website.
g2 = data.groupby(["Total Time Spent on Website in Minutes"]).size().sort_values(ascending=False).reset_index(name='Number of people')
print(g2)

#Getting the time which is been spent by maximum number of people.
print(g2.loc[0])
```

```python
#sns.displot(data, x = 'Total Time Spent on Website in Minutes', kind = 'kde')
sns.displot(data, x = 'Total Time Spent on Website in Minutes', kind = 'kde', aspect=10/6)
plt.xlabel('Total Time Spent on Website in Minutes',fontsize = 15)
plt.ylabel('Density',fontsize = 15)
plt.show()
```

```python
# Convert the column of ‘Converted’ from values 1/0 to yes/no
data['Converted'] = data['Converted'].map({1.0: 'yes', 0.0: 'No'})
data.head(10)
```

```python
print( 'Number of Customers whose lead has been successfully converted', data['Converted'].value_counts()['yes'] )
print( 'Number of Customers whose lead was not converted',data['Converted'].value_counts()['No'] )
```

```python
plt.figure(figsize = (3,4), dpi = 150)
sns.countplot(x='Converted',data=data)  
plt.xlabel('Converted',fontsize = 12)
plt.ylabel('Number of Customers',fontsize = 12)
plt.show() 
```

```python
plt.figure(figsize = (3,4), dpi = 150)
sns.boxplot(x='Converted',y='Total Time Spent on Website in Minutes',data=data) 
plt.xlabel('Converted',fontsize = 12)
plt.ylabel('Total time spent in minutes',fontsize = 12)
plt.show() 
```
